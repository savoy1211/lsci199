```python
from h_wordorder import *
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bigram_model_results_windows1to5_moby_dick", index_col=0, parse_dates=True)
```

```python
df.plot()
```

```python
# H[words] = H[word set] + H[order | word set]
# H[order | word set] = H[words] - H[word set] = D 
# 
```

```python
df['window_size'] = df.index + 1
plt.plot(df['window_size'], df['h_words'] - df['h_wordset'])
plt.ylabel("Order information (bits)")
plt.xlabel("Window size")
```

```python
df['entropy_rate'] = df['h_words'] / df['window_size']
plt.plot(df['window_size'], df['entropy_rate'])
plt.ylabel("nth-order word entropy rate (bits)")
plt.xlabel("n")
```

```python
df['unordered_entropy_rate'] = df['h_wordset'] / df['window_size']
plt.plot(df['window_size'], df['unordered_entropy_rate'])
plt.ylabel("nth-order unordered word entropy rate (bits)")
plt.xlabel("n")
```

```python
df['entropy_rate_difference'] = (df['h_words'] - df['h_wordset']) / df['window_size']
plt.plot(df['window_size'], df['entropy_rate_difference'])
plt.ylabel("nth-order word entropy rate difference (bits)")
plt.xlabel("n")
```
