```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
text='p'
# file="results/2gram/bigram_model_results_windows1to5_pride_and_prejudice"
file="NEW_2gram_alpha1_windows1to5_"+text
```

```python
df = pd.read_csv(file, index_col=0, parse_dates=True)
df['window_size'] = [1,2,3,4,5]
ax = df.plot(x="window_size", y=["h_words", "h_wordset"], title=text+" (bigram)")
ax.set(xlabel="Window Size", ylabel="Entropy")
ax.set_xticks(range(1,6,1))
plt.show()
```

```python
df['h_wordorder'] = df['h_words'] - df['h_wordset']
df.plot(x="window_size", y=['h_wordorder'], title="'Pride and Prejudice' word order (bigram)").set(xlabel="Window Size", ylabel="Entropy")
plt.show()
```

```python
df['entropy_rate'] = df['h_words'] / df['window_size']
df.plot(x="window_size", y=['entropy_rate'], title="'Pride and Prejudice' entropy rate (bigram)").set(xlabel="n", ylabel="nth-order word entropy rate (bits)")
```

```python
df['unordered_entropy_rate'] = df['h_wordset'] / df['window_size']
df.plot(x="window_size", y=['unordered_entropy_rate'], title="'Pride and Prejudice' unordered entropy rate (bigram)").set(xlabel="n", ylabel="nth-order unordered word entropy rate (bits)")
```

```python
df['entropy_rate_difference'] = (df['h_words'] - df['h_wordset']) / df['window_size']
df.plot(x="window_size", y=['entropy_rate_difference'], title="'Pride and Prejudice' entropy rate difference (bigram)").set(xlabel="n", ylabel="nth-order unordered word entropy rate difference (bits)")
```
