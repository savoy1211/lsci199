```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("gpt2_results_windows1to5_pride_and_prejudice", index_col=0, parse_dates=True)
df['window_size'] = [1,2,3,4,5]
df.plot(x="window_size", y=["h_words", "h_wordset"], title="'Pride and Prejudice' (gpt2)").set(xlabel="Window Size", ylabel="Entropy")
plt.show()

# df1 = pd.read_csv("csv/bigram_model_results_windows1to5_two_cities", index_col=0, parse_dates=True)
# df1.plot()
# df2 = pd.read_csv("csv/bigram_model_results_windows1to5_moby_dick", index_col=0, parse_dates=True)
# df2.plot()
# df3 = pd.read_csv("csv/bigram_model_results_windows1to5_hard_times", index_col=0, parse_dates=True)
# df3.plot()
```

```python
df['h_wordorder'] = df['h_words'] - df['h_wordset']
df.plot(x="window_size", y=['h_wordorder'], title="'Pride and Prejudice' word order (gpt2)").set(xlabel="Window Size", ylabel="Entropy")
plt.show()
```

```python
df['entropy_rate'] = df['h_words'] / df['window_size']
df.plot(x="window_size", y=['entropy_rate'], title="'Pride and Prejudice' entropy rate (gpt2)").set(xlabel="n", ylabel="nth-order word entropy rate (bits)")
```

```python
df['unordered_entropy_rate'] = df['h_words'] / df['window_size']
df.plot(x="window_size", y=['unordered_entropy_rate'], title="'Pride and Prejudice' unordered entropy rate (gpt2)").set(xlabel="n", ylabel="nth-order unordered word entropy rate (bits)")
```

```python
df['entropy_rate_difference'] = (df['h_words'] - df['h_wordset']) / df['window_size']
df.plot(x="window_size", y=['entropy_rate_difference'], title="'Pride and Prejudice' entropy rate difference (gpt2)").set(xlabel="n", ylabel="nth-order unordered word entropy rate difference (bits)")
```
